```python
import httplib2
import os
import sys
from pprint import pprint
from apiclient.discovery import build

channel_id = 'UC79hJz6y1EEiIkwfHOuWC4w'
youtube = build('youtube', 'v3', developerKey=api_key2)
youtube_link = 'https://www.youtube.com/watch?v='


# 채널의 구독자수, 업로드한 비디오 수
sub = youtube.channels().list(id = channel_id,
                              part='statistics').execute()
subscriberCount = sub['items'][0]['statistics']['subscriberCount']
videoCount = sub['items'][0]['statistics']['videoCount']
print('구독자 수: ', subscriberCount)
print('총 비디오 수: ', videoCount)


# 해당 채널의 모든 영상 가져오기
def get_channel_videos(channel_id):
    res = youtube.channels().list(id= channel_id,
                                  part='contentDetails').execute()     
    # 채널 메인페이지에서 업로드 누르기
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    videos = []
    next_page_token = None   
    
    while 1:
        # 업로드에 있는 비디오id 가져오기
        # 50개 제한
        res = youtube.playlistItems().list(playlistId = playlist_id,
                                           part = 'snippet',
                                           maxResults = 50,
                                           pageToken = next_page_token).execute()
        videos += res['items']
        
        # 50개 이후부턴 next page token으로 가져오기
        next_page_token = res.get('nextPageToken')  
        if next_page_token is None:
            break            
    return videos

videos = get_channel_videos(channel_id)
print('위 채널의 총 플레이리스트: ',len(videos))



for index, video in enumerate(videos, start=1):
    # 채널, 영상id, 날짜, 제목, 정보,url, 썸네일 가져오기
    ind = index
    channel = video['snippet']['channelId']
    video_id = video['snippet']['resourceId']['videoId']
    publishedAt = video['snippet']['publishedAt']
    title = video['snippet']['title']
    description = video['snippet']['description']
    thumbnail = video['snippet']['thumbnails']['high']['url']

                
#     d_items = ([("id", video[snippet]['resourceId']['videoId']),
#                ('index', )])
    print('index', "\t", ind, "\n"
         'channel',"\t", channel,"\n"
         'video_id',"\t", video_id,"\n"
         'publishedAt',"\t", publishedAt,"\n"
         'title',"\t", title,"\n"
         'description',"\t", description, "\n"
         'url',"\t", youtube_link + video_id, "\n"
         'thumbnail', "\t", thumbnail, "\n" "\n","\n","\n") 

    
'''
공영방송도 크롤링, comments, formatting
'''
```

```python
#비디오id 마다 statistics 가져오기
def get_video_stats(video_ids):
    stats = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(id = ','.join(video_ids[i:i+50]),
                                   part = 'statistics').execute()
        stats += res['items']
        print('stats',len(stats))
    return stats

stats = get_video_stats(video_ids)
len(stats)
print(stats)
```

```python
import pandas as pd

Counter = []
Channel_id = []
Video_id = []
Comment_id = []
Parent_id = []
PublishedAt = []
Name = []
Comment = []
Like_count = []
Reply_count = [] 


# video_id = '9j37E-mwgDI'
#비디오 id 마다 댓글, 대댓글 가져오기
#API 가 댓글, 대댓글의 dislike정보는 서포트 안함
def get_comments(video_id):
    next_page_token = None
    comments = []
    while 1:
        response = youtube.commentThreads().list(
            part = 'snippet',
            videoId = video_id,
            maxResults = 100, 
            textFormat = 'plainText',
            pageToken = next_page_token).execute()

        comments += response['items']

        next_page_token = response.get('nextPageToken')
        if next_page_token is None:
            break
    return comments


for video_id in video_ids:
    print(video_id)
    try:
        comment_extracter = get_comments(video_id) 
    except:
        print("This page has comments blocked")


for index, item in enumerate(comment_extracter, start=1):
    counter = len(comment_extracter)
    video_id = video_id
    parent_id = 0
    comment_id  = item['snippet']['topLevelComment']['id']
    publishedAt = item['snippet']['topLevelComment']['snippet']['publishedAt']
    name = item["snippet"]['topLevelComment']["snippet"]["authorDisplayName"]
    comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
    like_count = item['snippet']['topLevelComment']['snippet']['likeCount']
    reply_count = item["snippet"]['totalReplyCount']
    
    print('counter',"\t", counter, '\n'
      'video_id',"\t", video_id, '\n'
      'parent_id',"\t", parent_id, '\n'
      'comment_id',"\t", comment_id, '\n'
      'publishedAt',"\t", publishedAt, '\n'
      'name',"\t", name, '\n'
      'comment',"\t", comment, '\n',
      'like_count',"\t", like_count, '\n',
      'reply_count',"\t", reply_count, '\n','\n','\n')


    Counter.append(counter)
    Channel_id.append(channel_id)
    Video_id.append(video_id)
    Parent_id.append(parent_id)
    Comment_id.append(comment_id)
    PublishedAt.append(publishedAt)
    Name.append(name)
    Comment.append(comment)
    Like_count.append(like_count)
    Reply_count.append(reply_count)



    #대댓글 있으면 가져오기 (100개 제한)
    #날짜순으로밖에 못가져옴 (베댓 순으로 가져오고싶지만 ㅠㅠ)
    if reply_count > 0:
        replies = []
        parent = item["snippet"]['topLevelComment']["id"]
        response2 = youtube.comments().list(part='snippet', 
                                            maxResults=100, 
                                            parentId=parent,
                                            textFormat="plainText").execute()
        replies += response2['items']
        for i, item in enumerate(replies, start =1):
            counter = len(replies)
            parent_id = parent
            comment_id = item["id"]
            publishedAt = item["snippet"]['publishedAt']
            name = item["snippet"]["authorDisplayName"]
            comment = item["snippet"]["textDisplay"]
            like_count = item["snippet"]['likeCount']
            reply_count = 0
            
            print('r_counter',"\t", counter, '\n'
              'r_video_id',"\t", video_id, '\n'
              'r_parent_id',"\t", parent_id, '\n'
              'r_comment_id',"\t", comment_id, '\n'
              'r_publishedAt',"\t", publishedAt, '\n'
              'r_name',"\t", name, '\n'
              'r_comment',"\t", comment, '\n',
              'r_like_count',"\t", like_count, '\n',
              'r_reply_count',"\t", reply_count, '\n','\n','\n')
            

            Counter.append(counter)
            Video_id.append(video_id)
            Parent_id.append(parent_id)
            Comment_id.append(comment_id)
            PublishedAt.append(publishedAt)
            Name.append(name)
            Comment.append(comment)
            Like_count.append(like_count)
            Reply_count.append(reply_count)
            
```

```python
features = [Counter, Video_id, Parent_id, Comment_id, PublishedAt,
                   Name, Comment, Like_count, Reply_count]
```

```python
df = pd.DataFrame(features).transpose()
```

```python
df.columns = ['Counter', 'Video_id', 'Parent_id', 'Comment_id', 'PublishedAt',
                   'Name', 'Comment', 'Like_count', 'Reply_count']
```
